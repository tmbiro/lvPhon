```python
# Dependencies
import pandas as pd
import numpy as np

# Make sure you can see all output
pd.options.display.max_rows = 4000
```

```python
# Store filepath in a variable
df = pd.read_csv("Resources/lvPhon-MRI-Acc-dat.csv")
BNA = pd.read_csv("Resources/BNA_subregions.csv")
```

```python
# Check data
df.head()
```

```python
# Check atlas
BNA.head()
```

```python
# Rename ID Number columns in both dataframes
df = (
    df
    .rename(columns={
        'label':'Region_ID_Num'
        })
)

BNA = (
    BNA
    .rename(columns={
        'Label ID':'Region_ID_Num'
        })
)
```

```python
# Merge anatomical information into the participant dataset
df2 = pd.merge(df, BNA, on='Region_ID_Num', how='left')
df2.head()
```

```python
# Save as CSV
df2.to_csv("Resources/lvPhon-MRI-Acc-dat-v3.csv")
```
