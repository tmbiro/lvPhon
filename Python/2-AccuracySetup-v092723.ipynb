```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Store filepath in a variable
df = pd.read_csv("Resources/pvm_all_data.csv")

df.columns.tolist()
```

```python
#Make a dataframe to store accuracy scores
acc_scores = (
    df[[
    'PID',
    'Target',
    'Production',
    'Prod_Arpabet',
    'Word_ID',
    'Session_ID',
    'Prod_Word_N',
    'Prod_Phon_N',
    'Prod_Word_IPA',
    'Prod_Phon_IPA',
    'Prod_Phoneme_ID',
    'Target_Arpabet',
    'Target_N_Tot_Words',
    'Target_Word_IPA',
    'Target_N_Tot_Syllables',
    'Target_N_Tot_Phonemes',
    'Target_Phon_Arpabet',
    'Target_Phoneme_ID',
    'Target_Syll_Env',
    'Target_Word_Pos',
    'Syllable_NumID',
    'Target_Word_NumID',
    'Target_Con_Cluster',
    'Target_Clust_ID',
    'Target_Clus_Type',
    'Target_Clust_Phon_Pos',
    'Target_Clust_Phon_Env',
    'Target_Phon_IPA']]
    )
```

```python
# apply function to calculate voicing accuracy
acc_scores['Phon_Acc'] = df.apply(lambda x: 1 if x['Prod_Phoneme_ID'] == x['Target_Phoneme_ID'] else 0, axis=1)
acc_scores.head()
```

```python
# apply function to calculate voicing accuracy
acc_scores['Voicing_Acc'] = df.apply(lambda x: 1 if x['Prod_voice'] == x['Target_voice'] else 0, axis=1)
acc_scores.head()
```

```python
# apply function to calculate place accuracy
acc_scores['Place_Acc'] = df.apply(lambda x: 1 if x['Prod_Place_N'] == x['Target_Place_N'] else 0, axis=1)
acc_scores.head()
```

```python
# apply function to calculate manner accuracy
acc_scores['Manner_Acc'] = df.apply(lambda x: 1 if x['Prod_Manner_N'] == x['Target_Manner_N'] else 0, axis=1)
acc_scores.head()
```

```python
# apply function to calculate height accuracy
acc_scores['Height_Acc'] = df.apply(lambda x: 1 if x['Prod_Height_N'] == x['Target_Height_N'] else 0, axis=1)
acc_scores.head()
```

```python
# apply function to calculate frontness accuracy
acc_scores['Frontness_Acc'] = df.apply(lambda x: 1 if x['Prod_Frontness_N'] == x['Target_Frontness_N'] else 0, axis=1)
acc_scores.head()
```

```python
# apply function to calculate tenseness accuracy
acc_scores['Tenseness_Acc'] = df.apply(lambda x: 1 if x['Prod_tense'] == x['Target_tense'] else 0, axis=1)
acc_scores.head()
```

```python
# apply function to calculate roundness accuracy
acc_scores['Roundness_Acc'] = df.apply(lambda x: 1 if x['Prod_round'] == x['Target_round'] else 0, axis=1)
acc_scores.head()
```

```python
#Save accuracy scores
acc_scores.to_csv("Resources/acc_scores.csv", index=False, header=True)
```
